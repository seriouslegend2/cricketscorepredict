```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
import scipy.stats as stats

data = pd.read_csv('mth.csv')

X = data[['over', 'batting_team', 'bowling_team', 'innings', 'wickets']]
y = data['runs']

# converting strings to binary(true-1, else 0)
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), ['batting_team', 'bowling_team']),
        ('num', 'passthrough', ['over', 'innings', 'wickets'])
    ])

#converting categorical variables into numerical format
X_processed = preprocessor.fit_transform(X)

# Split the dataset into training (70%) and testing (30%)
X_train, X_test, y_train, y_test = train_test_split(X_processed, y, test_size=0.3, random_state=42)

# model weights (coefficients), theta=no.of features in x-train
theta = np.zeros(X_train.shape[1])

learning_rate = 0.01
epochs = 1000

def gradient_descent(X_train, y_train, theta, learning_rate, epochs):
    m = len(y_train)
    for _ in range(epochs):
        predictions = X_train.dot(theta)
        errors = predictions - y_train #differ b/w act and pred
        gradient = (1/m) * X_train.T.dot(errors)
        theta -= learning_rate * gradient
    return theta

theta = gradient_descent(X_train, y_train, theta, learning_rate, epochs)

# Make predictions on the test set
y_pred = X_test.dot(theta)

mse = mean_squared_error(y_test, y_pred) #low=better pred accuracy
mae = mean_absolute_error(y_test, y_pred)#sim to MSE but sensitive to outlayers
r2 = r2_score(y_test, y_pred)#closer to 1 better, vise versa

print(f"Trained Model Parameters: {theta}")
print(f"Mean Squared Error (MSE) on Test Set: {mse}")
print(f"Mean Absolute Error (MAE) on Test Set: {mae}")
print(f"R^2 Score on Test Set: {r2}")

# Plot Model Evaluation Metrics
metrics = {'MSE': mse, 'MAE': mae, 'R^2': r2}
plt.figure(figsize=(8, 6))
plt.bar(metrics.keys(), metrics.values(), color=['red', 'blue', 'green'])
plt.title('Model Evaluation Metrics')
plt.ylabel('Score')
plt.show()

residuals = y_test - y_pred

# Distribution of Residuals
plt.figure(figsize=(8, 6))
sns.histplot(residuals, kde=True, color='orange')
plt.title('Distribution of Residuals')
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.show()

# Residual Plot
plt.figure(figsize=(8, 6))
plt.scatter(y_pred, residuals, alpha=0.5, color='purple')
plt.axhline(y=0, color='black', linestyle='--')
plt.title('Residual Plot: Actual vs Predicted')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.show()

# Q-Q Plot of Residuals
plt.figure(figsize=(8, 6))
stats.probplot(residuals, dist="norm", plot=plt)
plt.title('Q-Q Plot of Residuals')
plt.show()

# Linear Regression Plot
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, color='blue', alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle='--')
plt.title('Linear Regression Plot: Actual vs Predicted')
plt.xlabel('Actual Runs')
plt.ylabel('Predicted Runs')
plt.show()

def predict_runs(over, batting_team, bowling_team, innings, wickets):
    # Create a dataframe for the given input
    input_data = pd.DataFrame({
        'over': [over],
        'batting_team': [batting_team],
        'bowling_team': [bowling_team],
        'innings': [innings],
        'wickets': [wickets]
    })
    
    input_processed = preprocessor.transform(input_data)
    
    predicted_runs = input_processed.dot(theta)
    
    print(f"Predicted Runs: {predicted_runs[0]}")
    return predicted_runs[0]
    
over = 15  # Example input
batting_team = 'Team A'  # Example input
bowling_team = 'Team B'  # Example input
innings = 1  # Example input
wickets = 7  # Example input

# Call the function to get the predicted runs
predict_runs(over, batting_team, bowling_team, innings, wickets)
```
