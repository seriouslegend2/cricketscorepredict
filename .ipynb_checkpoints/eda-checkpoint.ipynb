```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('./archive/match_data.csv')
df_info = pd.read_csv('./archive/match_info.csv')
```

## Univariate Analysis

### Numerical Columns

```python
# Runs distribution
sns.histplot(df['runs'], kde=True, color="blue", bins=30)
plt.title("Runs Distribution")
plt.xlabel("Runs")
plt.ylabel("Frequency")
plt.show()
```

```python
# Extras boxplot
sns.boxplot(x=df['extras'], color="orange")
plt.title("Distribution of Extras")
plt.show()
```

### Categorical Columns

```python
# Team performance frequency
df['batting_team'].value_counts().plot(kind='barh', figsize=(10, 6))
plt.title("Number of Matches Played by Each Team")
plt.xlabel("Frequency")
plt.ylabel("Teams")
plt.show()
```

```python
# Stadium distribution
sns.countplot(y="stadium", data=df, order=df['stadium'].value_counts().index, palette="viridis")
plt.title("Most Frequent Stadiums")
plt.xlabel("Frequency")
plt.ylabel("Stadiums")
plt.show()
```

## Bivariate Analysis

```python
# Runs by over
sns.lineplot(data=df.groupby('over')['runs'].sum().reset_index(), x='over', y='runs', marker='o')
plt.title("Runs Scored by Over")
plt.xlabel("Over")
plt.ylabel("Runs")
plt.show()
```

```python
# Runs scored by team
sns.boxplot(data=df, x="batting_team_short_name", y="runs", palette="coolwarm")
plt.title("Team-wise Runs Scored")
plt.xticks(rotation=45)
plt.show()
```

```python
# Wickets by team
sns.barplot(data=df.groupby('batting_team')['wickets'].sum().reset_index(), x='batting_team', y='wickets', palette="coolwarm")
plt.title("Total Wickets Lost by Each Team")
plt.xticks(rotation=90)
plt.show()
```

```python
# Match results based on toss decisions
sns.countplot(data=df_info, x="toss_decision", hue="winner", palette="mako")
plt.title("Impact of Toss Decision on Match Winners")
plt.xlabel("Toss Decision")
plt.ylabel("Frequency")
plt.xticks(rotation=0)
plt.show()
```

## MultiVariate Analysis

```python
# Performance in stadiums
stadium_runs = df.groupby(['venue'])['runs'].sum().sort_values(ascending=False).reset_index()
sns.barplot(data=stadium_runs.head(10), x="runs", y="venue", palette="viridis")
plt.title("Top Stadiums by Runs Scored")
plt.xlabel("Total Runs")
plt.ylabel("Stadium")
plt.show()
```

```python
# Performance in stadiums
stadium_runs = df.groupby(['venue'])['runs'].sum().sort_values(ascending=False).reset_index()
sns.barplot(data=stadium_runs.head(10), x="runs", y="venue", palette="viridis")
plt.title("Top Stadiums by Runs Scored")
plt.xlabel("Total Runs")
plt.ylabel("Stadium")
plt.show()
```

## Further Analysis

```python
# Most runs scored by a batsman
top_batsmen = df.groupby('striker')['runs_off_bat'].sum().sort_values(ascending=False).head(10)
sns.barplot(x=top_batsmen.values, y=top_batsmen.index, palette="coolwarm")
plt.title("Top 10 Batsmen by Runs")
plt.xlabel("Total Runs")
plt.ylabel("Batsman")
plt.show()
```

```python
# Most wickets taken by a bowler
top_bowlers = df.groupby('bowler')['is_wicket'].sum().sort_values(ascending=False).head(10)
sns.barplot(x=top_bowlers.values, y=top_bowlers.index, palette="mako")
plt.title("Top 10 Bowlers by Wickets")
plt.xlabel("Total Wickets")
plt.ylabel("Bowler")
plt.show()
```

```python
# Matches trend
matches_per_season = df_info['season'].value_counts().sort_index()
matches_per_season.plot(kind='bar', color="purple")
plt.title("Number of Matches Played Across Seasons")
plt.xlabel("Season")
plt.ylabel("Matches Played")
plt.show()
```
